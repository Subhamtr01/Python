```python
#Q1 Load the Sample 1 CSV File
import pandas as pd
sample1 = pd.read_csv(r"sample1_group7_035035_035019.csv", low_memory=False)
sample1
```

```python
for column_headers in sample1.columns: 
    print(column_headers)
```

```python
#Q2 Generate a Report on Feature-wise Missing Data in the Sample
missing_values = sample1.isnull()

# Using the sum() function to sum up the number of missing values in each column
feature_wise_missing_data = missing_values.sum()

# Print the report
print(feature_wise_missing_data)
```

```python
#Q3 Treat Missing Data in the Sample
#Q4 Recode Features with Text Data into Numeric Data in the Sample (Do Not Delete the Features with Text Data. Create New Features with Equivalent Numeric Data). Use the Recoded Features for all Analysis

recodes = {"False": 0, "True": 1}
recodes_type ={"A": 1, "B": 2, "C": 3}
sample1 = sample1.assign(new_colum_IsHoliday = sample1['IsHoliday'])
sample1 = sample1.assign(new_column_Type = sample1['Type'])
sample1["new_colum_IsHoliday"]=sample1["new_colum_IsHoliday"].apply(str)
sample1["new_column_Type"]=sample1["new_column_Type"].apply(str)

sample1["new_colum_IsHoliday"]=sample1["new_colum_IsHoliday"].map(recodes)
sample1["new_column_Type"]=sample1["new_column_Type"].map(recodes_type)

print(sample1["new_colum_IsHoliday"])
print(sample1["new_column_Type"])
sample1 = sample1.fillna(sample1.mean())
#sp1["Date"] = sp1["Date"].fillna('05/02/2010')
print(sample1)
```

```python
#Q5 Save the Processed Sample 1 as Sample 2 (sample1_xxx_yyy.csv) in form of CSV File in the Local Folder Location

sample1.to_csv("sample2_group07_035019_035035.csv")
```
